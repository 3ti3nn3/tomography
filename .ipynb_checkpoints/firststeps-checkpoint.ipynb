## Task 01: Creating uniformly distributed points on the Bloch sphere

```python
# general functions
def visualize_qubit(points=np.array([None]), states=np.array([None]), kind='point', angles=[-60, 30]):
    '''
    Depending on the parameters the function creates vectors or points on the Bloch sphere. 
    
    :param points: Nx3 array of expectations values (<sigma_x>, <sigma_y>, <sigma_z>)
    :param states: list of N Quatum objects 
    :param kind  : str which should be either "point" or "vector"
    :param angles: list of angles from which the Bloch sphere is seen
    '''
    b              = qt.Bloch()
    b.point_marker = 'o'
    b.point_color  = ['blue']
    b.vector_color = ['red']
    b.vector_width = 1
    b.point_size   = [10]
    b.view         = angles

    if np.all(points != None): 
        b.add_points(points)
        b.render()
    if np.all(states != None):
        b.add_states(states, kind=kind)
        b.render()
    b.show()
```

### Frist Approach: Linear distribution of angles

**Date: 25.10.22** <br>
**Idea:** Create a grid of angle parameteres $\phi$ and $\theta$. Use them in the representation of the following form: 

\begin{equation}
    |\Phi\rangle = \cos\bigg(\frac{\theta}{2}\bigg) |0\rangle + \exp(i\varphi) \sin\bigg(\frac{\theta}{2}\bigg) |1\rangle
\end{equation}

for $\phi\in [0, 2\pi), \theta\in [0, \pi)$. 

```python
def generate_linear(n_phi: int, n_theta: int):
    '''
    Generates linear distributed states,i.e. linear distributed angles in the allowed range.
    
    :param n_phi  : number of linear distributed polar points
    :param n_theta: number of linear distributed azimuth points
    :return: tuple of two arrays with angles
    '''
    phi             = np.linspace(0, 2*np.pi, n_phi)
    theta           = np.linspace(0, np.pi, n_theta)
    phi_v, theta_v  = np.meshgrid(phi, theta)
    
    return phi_v.flatten(), theta_v.flatten()
    

def angles_to_states(phi, theta):
    '''
    Takes angles and converts them into states in the compuational basis. 
    
    :param phi  : array or float of polar angles
    :param theta: array or float of azimuth angles
    return: Qobj in computational basis
    '''
    up     = qt.basis(2, 0)
    down   = qt.basis(2, 1)
    
    return np.cos(theta/2).tolist()*up + ( np.sin(theta/2) * np.exp(1j*phi) ).tolist()*down
```

```python
phi, theta = generate_linear(50, 50)
visualize_qubit(states=angles_to_states(phi, theta))
```

### Second Approach: Create States by with the the concept of inverse transform sampling
https://en.wikipedia.org/wiki/Inverse_transform_sampling

**Date: 26.10.2022** <br>
**Idea:** The proof of concept of inverse transfrom sampling is given by the probability integral transform statement (https://en.wikipedia.org/wiki/Probability_integral_transform).

```python
def generate_pure_uniform(N: int):
    '''
    Generates data in compliance with transforming surface element. Data is uniformly smapled according to
    the concept of Inverser Transform Sampling.
    
    :param N: number of uniformly distributed points
    :return: an array both for the polar and the azimuth angle of lenght N  
    '''
    x     = np.random.uniform(0, 1, size=N)
    theta = np.arccos(1-2*x)
    phi   = np.random.uniform(-np.pi, np.pi, size=N)
    
    return phi, theta
```

### Third Approach: Use the probability distribution from before and create states with unitary transformation

**Date: 27.10.2022** <br>
**Idea:** Instead of using the concept of the first approach we want to define a unitary operator to be able to generalize to more than one qubit. Therefore, we define three rotation $R_x(\alpha)$, $R_y(\theta)$ and $R_z(\phi)$. These matrices rotate a given state $\rho$ around the cartesian axes $x$, $y$ and $z$:

\begin{align}
    R_x(\alpha) &= 
    \begin{pmatrix}
            \cos\big(\frac{\alpha}{2}\big) & -i\sin\big(\frac{\alpha}{2}\big) \\
            -i\sin\big(\frac{\alpha}{2}\big) & \cos\big(\frac{\alpha}{2}\big) \\
    \end{pmatrix} \\
    R_y(\theta) &= 
    \begin{pmatrix}
            \cos\big(\frac{\theta}{2}\big) & -\sin\big(\frac{\theta}{2}\big) \\
            \sin\big(\frac{\theta}{2}\big) & \cos\big(\frac{\theta}{2}\big) \\
    \end{pmatrix} \\
    R_z(\varphi) &= 
    \begin{pmatrix}
            \exp\big(i\frac{\varphi}{2}\big) & 0 \\
            0 & \exp\big(-i\frac{\varphi}{2}\big) \\
    \end{pmatrix}
\end{align}

Starting with a randomly choosen state $\rho_0 = \ket{\Psi}\bra{\Psi}$ one can build with the randomly sampled angles $\phi$ and $\theta$ a uniform distribution of states on the Bloch sphere via: 

\begin{equation}
    \rho = R(\varphi, \theta) \rho_0 R(\varphi, \theta)^{\dagger}
\end{equation}

where $R(\phi, \theta) = R_z(\phi)R_y(\theta)$. One has to keep in mind that depending on the starting state $\rho_0$ one should choose the order of ratations, i.e. rotating the state 

\begin{equation}
    \rho_0 = 
        \begin{pmatrix} 
            1 & 0 \\
            0 & 0 \\
        \end{pmatrix}
\end{equation}

first around the $z$-axis is not useful. 

```python
# Pauli matrices
sigmae = np.eye(2)
sigmax = np.array([[0, 1],
                   [1, 0]])
sigmay = np.array([[0, 1j],
                   [-1j, 0]])
sigmaz = np.array([[1, 0],
                   [0, -1]])
sigma  = np.array([sigmax, sigmay, sigmaz])

# rotation matrices 
def Rx(alpha):
    return np.transpose([[np.cos(alpha/2), -1j*np.sin(alpha/2)], 
                         [-1j*np.sin(alpha/2), np.cos(alpha/2)]], axes=[2, 0, 1])
def Ry(theta): 
    return np.transpose([[np.cos(theta/2), -np.sin(theta/2)],
                         [np.sin(theta/2), np.cos(theta/2)]], axes=[2, 0, 1])
def Rz(phi):
    return np.transpose([[np.exp(1j*phi/2, dtype=complex), np.zeros(len(phi), dtype=complex)],
                         [np.zeros(len(phi), dtype=complex), np.exp(-1j*phi/2, dtype=complex)]], axes=[2, 0, 1])
```

```python
def expect(operator: np.array, rho: np.array):
    '''
    Calculates the expectation value of an operator and a state.
    
    :param operator: operator
    :param rho     : state
    :return: expectation value
    '''
    return np.trace(operator@rho, axis1=-2, axis2=-1)

def angles_to_density(phi, theta):
    '''
    Takes polar and azimuth angles and builds a vector of expecation values in cartesian coordinates. 
    
    :param phi  : array or float of polar angle 
    :param theta: array or float of azimuth angle 
    return: 3xlen(phi)-dimensional array of expectations values in cartesian coordinates
    '''
    R   = Rz(phi)@Ry(theta)
    RH  = np.transpose(np.conjugate(R), axes=[0, 2, 1])
    rho = R@np.array([[1, 0], [0, 0]])@RH
    
    return rho
```

```python
phi, theta = generate_pure_uniform(1000)
rho        = angles_to_density(phi, theta)

visualize_qubit(points=np.array([expect(sigmax, rho), expect(sigmay, rho), expect(sigmaz, rho)]))
```

```python
def check_purity(rhos: np.array, prec=1e-14):
    '''
    Check purity of a given array of density matrix.
    
    :param rhos: array of density matrices
    :param prec: precision of the purity comparison
    :return: boolean whether given density matrices are pure or not
    '''
    purity = np.trace(rhos@rhos, axis1=-2, axis2=-1, dtype=complex)

    if np.all(np.abs(np.imag(purity)) < prec) and np.all(np.abs(purity-1) < prec):
        return True
    else: 
        shadow = np.logical_or(np.all(np.abs(np.imag(purity)) >= prec), np.all(np.abs(purity-1) >= prec))
        print('The follwing {0} states are not pure:\n'.format(np.sum(shadow)), rhos[shadow])
        print('Purity:\n', purity[shadow])
        return False
```

```python
def expectation_distribution(rho, n_bins=10):
    
    N = len(rho)
    
    expect_x = expect(sigmax, rho) 
    expect_y = expect(sigmay, rho)
    expect_z = expect(sigmaz, rho)

    mu_x = np.sum(expect_x)/N
    mu_y = np.sum(expect_y)/N
    mu_z = np.sum(expect_z)/N
    
    # plot
    fig, axs = plt.subplots(1, 3, figsize=(30, 6), sharey=True)
    axs[0].set_ylabel('counts')

    fig.suptitle('Expectation Values of Approach two')

    n_x, _, _ = axs[0].hist(expect_x, bins=n_bins)
    n_y, _, _ = axs[1].hist(expect_y, bins=n_bins)
    n_z, _, _ = axs[2].hist(expect_z, bins=n_bins)

    axs[0].vlines(mu_x, 0, np.max(n_x), color='red')
    axs[1].vlines(mu_y, 0, np.max(n_y), color='red')
    axs[2].vlines(mu_z, 0, np.max(n_z), color='red')

    axs[0].text(0.075, np.max(n_x)/2, '{0:.2e}'.format(mu_x), color='red')
    axs[1].text(0.075, np.max(n_x)/2, '{0:.2e}'.format(mu_y), color='red')
    axs[2].text(0.075, np.max(n_x)/2, '{0:.2e}'.format(mu_z), color='red')

    axs[0].set_title(r'Expectation value $<\sigma_x>$')
    axs[1].set_title(r'Expectation value $<\sigma_y>$')
    axs[2].set_title(r'Expectation value $<\sigma_z>$')

    axs[0].set_xlabel(r'$<\sigma_x>$')
    axs[1].set_xlabel(r'$<\sigma_y>$')
    axs[2].set_xlabel(r'$<\sigma_z>$')

    axs[0].set_xlim(-1, 1)
    axs[1].set_xlim(-1, 1)
    axs[2].set_xlim(-1, 1)

    plt.show()
```

### Fourth Approach: Using arbitray unitary matrices

```python
def unitary_to_density(dim: int, N: int):
    '''
    Creates uniformly distributed states on the Bloch sphere by using arbitrary unitary operators.
    
    :param dim: dimension 
    :param N  : number of random density matrices
    :return: array of N uniformly distributed density matrices
    '''
    U   = unitary_group.rvs(dim, size=N)
    UH  = np.transpose(np.conjugate(U), axes=[0, 2, 1])
    rho = U@np.array([[1, 0], [0, 0]])@UH
    
    return rho
```

### Fith Approach: Angles direct to density matrices

```python
def direct_to_density(phi: np.array, theta: np.array):
    
    Psi = np.array([np.cos(theta/2), np.sin(theta/2)*np.exp(1j*phi)]).T
    
    return np.einsum('nk,nj->nkj', Psi, np.conjugate(Psi))
```

## Task 02: Creating uniformly distributed mixed states

### First Approach: Bloch vector

**Date: 29.09.2022** <br>
**Idea: Use the the representation of mixed states givien by the Bloch vector and create uniform distributed radi, polar and azimuth angle.**

\begin{equation}
    \rho = \frac{1}{2} \big(\mathbb{1} + \vec{n}\vec{\sigma} \big)
\end{equation}

Here, $\vec{n}$ is called the Bloch vector.

```python
def generate_mixed_uniform(N: int):
    '''
    Generates uniformly distributed Bloch parameters for mixed states.
    
    :param N: number of samples
    :return: tuple of arrays of radii, polar angles and azimuth angles
    '''
    x     = np.random.uniform(0, 1, size=N)
    y     = -1*np.random.uniform(-1, 0, size=N)
    
    theta = np.arccos(1-2*x)
    phi   = np.random.uniform(-np.pi, np.pi, size=N)
    r     = np.power(y, 1/3)
    
    return r, phi, theta
```

```python
def blochvector_to_density(r: np.array, phi: np.array, theta: np.array):
    '''
    Builds mixed states via the Bloch representation.
    
    :param r    : array of radii
    :param phi  : array of polar angles
    :param theta: array of azimuth angles
    :return: array of mixed states
    '''
    n = r*np.array([np.cos(phi)*np.sin(theta),
                    np.sin(phi)*np.sin(theta),
                    np.cos(theta)])
    
    if len(n.shape)==1:
        return 1/2 * (sigmae + np.einsum('k,klm->lm', n, sigma))
    else:
        return 1/2 * (sigmae + np.einsum('kn,klm->nlm', n, sigma))
```

```python
r, phi, theta = generate_mixed_uniform(10000)
rho           = blochvector_to_density(r, phi, theta)

visualize_qubit(points=np.array([expect(sigmax, rho), expect(sigmay, rho), expect(sigmaz, rho)]))
```

### Second approach: Arbitrary hermititan

**Idea: Create a gaussian distributed matrix and create a hermitian.**

Creating a gaussian arbitrary complex matrix $A$ and perform the following operation: 

\begin{equation}
    \rho = \frac{AA^{\dagger}}{tr(AA^{\dagger})}
\end{equation}

This process creates uniformly distributed mixed states according to Hilbert-Schmidt distance measure. 

```python
def hermitian_to_density(dim: int, N: int):
    A  = np.random.normal(size=[N, dim, dim]) + 1j*np.random.normal(size=[N, dim, dim])
    AH = np.transpose(np.conjugate(A), axes=[0, 2, 1])
    
    B   = A@AH
    rho = np.multiply(1/np.trace(B, axis1=-2, axis2=-1)[:, None, None].repeat(2, axis=1).repeat(2, axis=2), B)
    
    return rho
```

```python
rho = hermitian_to_density(2, 1000)
visualize_qubit(points=np.array([expect(sigmax, rho), expect(sigmay, rho), expect(sigmaz, rho)]))
```

### Speed test of different methods for creating mixed and pure states

```python
def speed_measure(func, *args, iterations=10):
    '''
    Measures the speed of the given function.
    
    :param func      : function which speed should be measured
    :param *args     : arguments of test function
    :param iterations: number of iterations the speed is measured
    :return: time the test function needed
    '''
    t0 = np.empty(iterations, dtype=float)
    t1 = np.empty(iterations, dtype=float)
    
    for i in range(iterations):
        t0[i] = time.time()
        func(*args)
        t1[i] = time.time()
    
    return np.mean(t1-t0)


def compare_speed(iterations=10, **kwargs):
    '''
    Compares a arbitraty number of functions.
    
    :param iterations: number of iterations the test function is tested
    :param **kwargs  : dictionary like objekt of the form "name = (func, list of parameters)"
    :return: dictionaries of times each test function needed 
    '''
    t_mean = {}
    
    for key, value in kwargs.items():
        t           = speed_measure(value[0], *value[1], iterations=iterations)
        t_mean[key] = t
        
    return t_mean
```

## Task 03: Linear Inversion

**Date: 02.10.22** <br>

```python
def check_povm(M: np.array):
    
    # hermitian
    MH = np.transpose(np.conjugate(M), axis1=-2, axis2=-1)
    bool_herm = M == MH
    
    # completeness
    prec  = 1e-15
    M_sum = np.sum(M, axis=0)
    bool_compl = np.any(np.abs(np.real(M_sum) - np.eye(len(M[0]))) < prec) and np.any(np.imag(M_sum) < prec)
    
    # positivity
    evalues, _ = np.linalg.eig(M)
    bool_pos   = np.all(evalues >= 0)
    
    return bool_herm and bool_compl and bool_pos
```

Remember the formalism of linear inversions for a general POVM $\{M_i\}_i$. This POVM is expected to be complete. Thus, we can expand the density matrix in terms of the POVM with coefficients $c_i \in \mathbb{C}$: 

\begin{equation}
    \rho = c_j M_j
\end{equation}

\begin{equation}
    \rho = c_j M_j = p_i tr(M_jM_i)^{-1} M_j = p_i T_{ij}^{-1} M_j
\end{equation}

```python
def linear_inversion(M: np.array, n):
    p     = np.mean(n, axis=1)
    T     = np.einsum('alk,bkl->ab', M, M)
    T_inv = np.linalg.inv(T)
    
    return np.einsum('i,ji,jnk->nk', p, T_inv, M)
```

Check for corectness. Consider the Pauli's $\{\mathbb{1}, \sigma_x, \sigma_y, \sigma_z\}$ and the measurement $\langle\sigma_x\rangle = \langle\sigma_y\rangle = \langle\sigma_z\rangle = 1$.

## Task 04: Maximum Likelihood Estimator

**04.10.2022**

The maximum likelihood method seeks the estimator which maximazises the likelihood function given by the following: 

\begin{equation}
    \mathcal{L}(\{\sigma_i\}_{i \in I}) = \prod_{i \in I} Tr(\rho\sigma_i)
\end{equation}

\begin{equation}
    \rho = \arg\max_{\rho} \mathcal{L}(\{\sigma_i\}_{i\in I})
\end{equation}

We numerically solve this optimization problem with the iterative MLE scheme: 

\begin{align}
    \rho_0 &= \mathbb{1} \\
    R(\rho) &= \sum_{i\in I_n}\frac{|\sigma_i\rangle\langle{\sigma_i|}}{tr(\rho_{n+1}|\sigma_i\rangle\langle|\sigma_i|)} \\
    \rho_{n+1} &= \mathcal{N}R(\rho_n)\rho_nR(\rho_n)
\end{align}

```python
eigenstate = {}
eigenstate[(1, +1)] = 1/np.sqrt(2)*np.array([1, 1])
eigenstate[(1, -1)] = 1/np.sqrt(2)*np.array([1, -1])
eigenstate[(2, +1)] = 1/np.sqrt(2)*np.array([1, 1j])
eigenstate[(2, -1)] = 1/np.sqrt(2)*np.array([1, -1j])
eigenstate[(3, +1)] = np.array([1, 0])
eigenstate[(3, -1)] = np.array([0, 1])


def mle_iterative_one(rho, R_arr):
    '''
    Performs one step of iterative MLE according to given measurement. 
    
    :param rho  : current numerical step
    :param R_arr: array of eigenstate matrices, is provided here to pretend recalculating each time
    :param d    : tuple of current measurement, first entry represents the measured axis
        second one is the observed eigenvalue
    :retur: the updated density matrix
    '''
    
    p_arr  = np.einsum('ik,nki->n', rho, R_arr)
    R      = np.einsum('n,nij->ij', 1/p_arr, R_arr)
    update = R@rho@R
    
    return update/np.trace(update) 


def mle_iterative(D: list, N: int): 
    '''
    Performs iterative MLE. 
    
    :param D: measured data set
    :return: iterative MLE estimate
    '''
    I     = len(D)
    rho   = np.eye(2)/2
    R_arr = np.empty((I, 2, 2), dtype=complex)
    
    for i in range(I):
        estate   = eigenstate[D[i]]
        R_arr[i] = np.tensordot(estate, np.conjugate(estate), axes=0)
    
    for i in range(N):
        rho = mle_iterative_one(rho, R_arr)

    return rho
```

```python
d1 = (1, 1)
d2 = (2, 1)
d3 = (1, 1)

D1 = [d1, d2, d3]

rho_1 = mle_iterative(D1, 1000)

print(rho_1)
```
